```python
import numpy as np
import matplotlib.pyplot as plt

def k_means(X, k, max_iters=100):
    """
    K-means clustering algorithm.

    Parameters:
    - X: Data matrix (each row is a data point).
    - k: Number of clusters.
    - max_iters: Maximum number of iterations.

    Returns:
    - centroids: Final cluster centers.
    - labels: Cluster assignments for each data point.
    """

    # Initialize centroids randomly
    centroids = X[np.random.choice(X.shape[0], k, replace=False)]

    for _ in range(max_iters):
        # Assign each data point to the nearest centroid
        distances = np.linalg.norm(X[:, np.newaxis, :] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        # Update centroids based on the mean of assigned points
        for i in range(k):
            centroids[i] = np.mean(X[labels == i], axis=0)

    return centroids, labels

# Generate random data (replace this with your own dataset)
np.random.seed(42)
X = np.concatenate([np.random.randn(50, 2) + [5, 5], np.random.randn(50, 2) + [10, 10]])

# Apply K-means with k=2
k = 2
centroids, labels = k_means(X, k)

# Visualize the clustered data
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.8)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

```
