```python
import pandas as pd
import numpy as np

def predict(w, b, X):
    return np.dot(X, w) + b

def gradient_descent(X, y, m, h):
    alpha = 0.001
    iterations = h
    b_curr = 0
    w_curr = np.zeros(X.shape[1])

    for i in range(iterations):
        y_cap = predict(w_curr, b_curr, X)

        slb = (2/m) * np.sum(y_cap - y)
        slw = (2/m) * np.dot((y_cap - y).T, X)

        w_curr = w_curr - alpha * slw
        b_curr = b_curr - alpha * slb

        print("\nIteration:", i+1)
        print("W_CURR =", w_curr, "\nB_CURR =", b_curr)

    return w_curr, b_curr

# Load your data (make sure to adjust the file path)
read = pd.read_csv('insurance.csv')
data = read.iloc[1:31]
n1 = data.iloc[:, 2:4]  # Assuming columns 2 and 3 are features
n2 = data.iloc[:, 6:7]   # Assuming column 6 is the target variable

X = np.array(n1)
y = np.array(n2).flatten()  # Flatten the target variable to a 1D array
m = len(X)

# Add a column of ones to X for the bias term
X_b = np.c_[np.ones((m, 1)), X]

# Set the number of iterations
h = 9

# Perform gradient descent
w_final, b_final = gradient_descent(X_b, y, m, h)

# Make predictions on the test set
n3 = data.iloc[21:31, 2:4]  # Assuming columns 2 and 3 are features in the test set
X_test = np.array(n3)
X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
y_pred = predict(w_final, b_final, X_test_b)

print("\nPredictions on the test set:", y_pred)
```
